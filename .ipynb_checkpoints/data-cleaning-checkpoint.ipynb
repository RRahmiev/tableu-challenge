```python
import pandas as pd
```

```python
# Loading first dataset - January bicycle data
df_jan = pd.read_csv('2019-01-data.csv')
df_jan.head()
```

```python
# Loading other datasets - April, July and October bicycle data
df_apr = pd.read_csv('2019-04-data.csv')
df_jul = pd.read_csv('2019-07-data.csv')
df_oct = pd.read_csv('2019-10-data.csv')
```

```python
# Combining data into one dataframe using the Concat method as all the datasets share the same column names and formatting
frames = [df_jan, df_apr, df_jul, df_oct]
data_df = pd.concat(frames)

# Printing combined dataframe
data_df.head()
```

```python
# Checking data types of each column
data_df.dtypes
```

### Checking for any missing/NaN values in the data

```python
check_nan = data_df.isnull().values.any()
 
# Printing the result
print(check_nan)
```

```python
# Renaming column names
data_df = data_df.rename({'tripduration': 'trip duration', 'starttime': 'start time',
                         'stoptime':'stop time', 'bikeid':'bike id', 'usertype':'user type'}, axis=1)
# Output:
data_df.head()
```

```python
# Replace numbers in column 'gender'
data_df = data_df.replace({'gender' : { 1 : 'male', 2 : 'female', 0 : 'unknown'}})
data_df.head()
```

### Saving final dataframe as a csv

```python
data_df.to_csv('citi-bike-data.csv')
```
